```python
def obtener_datos_twitter(artista):
    opciones = webdriver.ChromeOptions()
    opciones.add_argument("--start-maximized") 
    driver = webdriver.Chrome(PATH, options=opciones)
    URL = 'https://www.google.es/search?q='
    busqueda = artista + ' twitter official'

    # Abre página de Google y busca "artista + twitter official"
    driver.get(URL+busqueda)
    driver.find_element(By.XPATH, '//*[@id="W0wltc"]').click()  # rechaza cookies

    # Hacer clic en el enlace de Twitter del artista
    caja = driver.find_element(By.CSS_SELECTOR, 'div:nth-child(1) > div > div > div.Z26q7c.UK95Uc.jGGQ5e > div > a')
    caja.click()
    
    time.sleep(12) # Le doy un tiempo para que aparezca una ventana de notificaciones twitter que hace falta rechazar
    
    notificaciones = driver.find_element(By.CSS_SELECTOR, 'div:nth-child(2) > div > div > div > div > div > div.css-1dbjc4n.r-1awozwy.r-1kihuf0.r-18u37iz.r-1pi2tsx.r-1777fci.r-1pjcn9w.r-xr3zp9.r-1xcajam.r-ipm5af.r-g6jmlv > div.css-1dbjc4n.r-14lw9ot.r-1867qdf.r-1jgb5lz.r-pm9dpa.r-1ye8kvj.r-1rnoaur.r-13qz1uu > div > div.css-1dbjc4n.r-1awozwy.r-16y2uox > div > div.css-1dbjc4n.r-98ikmy.r-hvns9x > div.css-1dbjc4n.r-13qz1uu > div.css-18t94o4.css-1dbjc4n.r-1niwhzg.r-1ets6dv.r-sdzlij.r-1phboty.r-rs99b7.r-1wzrnnt.r-19yznuf.r-64el8z.r-1ny4l3l.r-1dye5f7.r-o7ynqc.r-6416eg.r-lrvibr > div > span > span')
    notificaciones.click()
    
    cookis = driver.find_element(By.XPATH, '//*[@id="layers"]/div/div[2]/div/div/div/div[2]/div[2]/div')
    cookis.click()
    
    # Obtenemos la cantidad de tweets del artista
    tweets = driver.find_element(By.XPATH, '//*[@id="react-root"]/div/div/div[2]/main/div/div/div/div/div/div[1]/div[1]/div/div/div/div/div/div[2]/div/div').text

    return [tweets]
```

```python
tweets_artistas = []

# Recorremos cada elemento de la columna Artist Name y aplicamos nuestra función 
for nombre_artista in df['Artist Name']:
    try:
        resultado = obtener_datos_twitter(nombre_artista)
        tweets_artistas.append(resultado)
    except:
        tweets_artistas.append(0)
```

```python
tweets_artistas
```

```python
# Ahora pasamos el resultado del scrappeo a una sola lista

nueva_lista_tweets = []
for sublista in tweets_artistas:
    if isinstance(sublista, list):
        nueva_lista_tweets.extend(sublista)
    else:
        nueva_lista_tweets.append(sublista)

print(nueva_lista_tweets)
```

```python
# Incorporamos esa lista como una columna de nuestro DataFrame

df['Tweets'] = nueva_lista_tweets
df
```

```python
# Ahora hacemos lo mismo con los seguidores de TikTok

def obtener_datos_tik(artista):
    PATH = "/ruta/al/chromedriver"
    opciones = webdriver.ChromeOptions()
    opciones.add_argument("--start-maximized") 
    opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
    opciones.add_experimental_option('useAutomationExtension', False)
    driver = webdriver.Chrome(PATH, options=opciones)
    URL = 'https://www.google.es/search?q='
    busqueda = artista + ' TikTok oficial'

    # Abre página de Google y busca "artista TikTok oficial"
    driver.get(URL+busqueda)
    driver.find_element(By.XPATH, '//*[@id="W0wltc"]').click()  # rechaza cookies

    # Hace clic en el enlace de tiktok del artista
    caja_tik = driver.find_element(By.CSS_SELECTOR, '#rso > div:nth-child(1) > div > div > div.rULfzc > div > a > h3')
    caja_tik.click()
    
    # Obtiene el texto con los seguidores de Tiktok
    followers_tik = driver.find_element(By.XPATH, '//*[@id="main-content-others_homepage"]/div/div[1]/h3/div[2]/strong').text

    # Cierra el navegador
    driver.quit()

    return [followers_tik]
```

```python
tiks_artistas = []

# Recorremos cada elemento de la columna Artist Name y aplicamos nuestra función
for nombre_artista in df['Artist Name']:
    try:
        resultado = obtener_datos_tik(nombre_artista)
        tiks_artistas.append(resultado)
    except:
        tiks_artistas.append(0)
```

```python
# Ahora pasamos el resultado del scrappeo a una sola lista

nueva_lista_tiks = []
for sublista in tiks_artistas:
    if isinstance(sublista, list):
        nueva_lista_tiks.extend(sublista)
    else:
        nueva_lista_tiks.append(sublista)

print(nueva_lista_tiks)
```

```python
# Incorporamos esa lista como una columna de nuestro DataFrame

df['TikTok Followers'] = nueva_lista_tiks
df
```
